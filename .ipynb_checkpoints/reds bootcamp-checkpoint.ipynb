```python
# Create a new column 'is_2_strikes' and assign the boolean result of the comparison 'strikes == 2' to it
df['is_2_strikes'] = df['strikes'] == 2
df.head()
```

```python
# Create a new column 'is_high_velo' and assign the boolean result of the comparison 'release_speed > 94' to it
df['is_high_velo'] = df['release_speed'] > 94
df.head()
```

```python
# Create a function to check the conditions for the new column
def is_strike_in_heart(row):
    description = row['description']
    attack_zone = row['attack_zone']
    
    # Check if the pitch is a called or swinging strike, not a foul
    if description in ['called_strike', 'swinging_strike'] and attack_zone == 'heart':
        return True
    else:
        return False

# Apply the function to create a new column 'is_strike_heart'
df['is_strike_heart'] = df.apply(is_strike_in_heart, axis=1)

df.head()
```

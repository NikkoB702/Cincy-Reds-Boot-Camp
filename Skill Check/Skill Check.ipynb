```python
import pandas as pd
```

```python
file_path = "C:/Users/Home/Cincy-Reds-Boot-Camp/Skill Check/bootcamp-skill-check - data.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# Add "home_score" and "away_score" together to form a new "total_score" column
df['total_score'] = df['home_score'] + df['away_score']

# Print the updated DataFrame
print(df)
```

```python
# Calculate the average release speed
average_speed = df['release_speed'].mean()

# Create a new column "release_speed_normalized" by dividing "release_speed" by the average speed
df['release_speed_normalized'] = df['release_speed'] / average_speed

# Print the updated DataFrame
print(df)
```

```python
# Make a new column "heart_shadow" that indicates if "attack_zone" is the heart or shadow
df['heart_shadow'] = df['attack_zone'].apply(lambda x: 1 if x in ['heart', 'shadow'] else 0)

# Make a new column "is_reds_pitcher" based on conditions for Cincinnati Reds pitching
df['is_reds_pitcher'] = df.apply(lambda row: 1 if (row['home_team'] == 'CIN' and row['inning_topbot'] == 'top') or (row['away_team'] == 'CIN' and row['inning_topbot'] == 'bot') else 0, axis=1)

# Print the updated DataFrame
print(df)
```

```python
# Calculate the mean of the "release_speed" column
mean_release_speed = df['release_speed'].mean()
print("Mean release speed:", mean_release_speed)

# Calculate the standard deviation of the "release_speed" column
std_release_speed = df['release_speed'].std()
print("Standard deviation:", std_release_speed)

# Calculate the difference between the mean and median of the "release_speed" column
median_release_speed = df['release_speed'].median()
mean_median_difference = mean_release_speed - median_release_speed
print("Difference between mean and median:", mean_median_difference)
```

```python
# Save the DataFrame to a CSV file
df.to_csv('skill_check_stats.csv', index=False)
```
